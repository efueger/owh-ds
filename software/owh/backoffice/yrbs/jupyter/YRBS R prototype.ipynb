```r
dim(D)
yrbs = svydesign(ids=~PSU,
                 strata=~STRATUM,
                 weights=~WEIGHT,
                 data=D)
# Stratified 1 - level Cluster Sampling design (with replacement)
# With (54) clusters.
# svydesign(ids = ~PSU, strata = ~STRATUM, weights = ~WEIGHT, data = D)

svymean(~QN8,yrbs)
svymean(~QN8,subset(yrbs,Sex=='Female'))
svymean(~QN8,subset(yrbs,Sex=='Male'))
svymean(~QN8,subset(yrbs,Race=='Asian'))
svymean(~QN8,subset(yrbs,Sex=='Male' & Race=='Asian'))
svymean(~QN8,subset(yrbs,Sex=='Female' & Race=='White'))
svymean(~QN8,subset(yrbs,Race=='Hispanic/Latino'))
```

```r
# Model question QN8="Rarely or never wore a bicycle helmet"
Q8model <- svyglm(QN8==1 ~ 0+Sex*Grade+Sex*Race, 
                  design=yrbs,
                  family=quasibinomial())
```

```r
inv.logit = function(x) exp(x) / (1+exp(x))
```

```r
#22 x 22
V = vcov(Q8model)
# 71 x 71
VV = L %*% V %*% t(L)

sev = sqrt(diag(VV))
```

```r
m = c(L %*% coef(Q8model))
MarginPred = inv.logit(cbind(mean=m,
                             lower=m-qnorm(0.975)*sev,
                             upper=m+qnorm(0.975)*sev))
MarginPred = as.data.frame(MarginPred)
# 71 x 7
MarginPred = cbind(L1,MarginPred,Count=nn)

```

```r
P = as.data.frame(predict(Q8model,newdata=G))
# 64 x 2
se = P$SE
link = P$link
PP = inv.logit(cbind(mean=link,
                     lower=link-qnorm(0.975)*se,
                     upper=link+qnorm(0.975)*se))
# 64 x 3
PP = as.data.frame(PP)
Pred = cbind(G,PP,Count=H$Freq)
# 135 x 7
Pred = rbind(Pred,MarginPred)
```
